```python
import pandas as pd
import matplotlib as plt

# file path copy
df = pd.read_csv('/content/sample_data/california_housing_train.csv')

# Read data
df.head(10)

```

```python
# 1. Data structure Overview
print("---- Data Info (df.info())----")
df.info()

# 2. Summary statistics of Numeric data
print("n\--- Summary of main statistics (df.describe())---")
df.describe()

# 3. Check the unique values and their frequencies for a specific column
# print("n\--- Ocean proximity and count---")
# print(df['ocean_proximity'].value_counts())
```

```python
# Create a new DataFrame containing only rows where 'median_income' is greater than 8
high_income_df = df[df['median_income']> 8]

# Check the first 5 rows of the filtered data
print(high_income_df.head(5))
```

```python
import matplotlib.pyplot as plt

# Set the figure size to ensure the plot is clearly visible in Colab
plt.figure(figsize=(10, 5))

# Plot a histogram for the 'median_house_value' column
# bins=50 specifies the number of bins (intervals) for the histogram
df['median_house_value'].hist(bins=50)

# Set the title and axis labels for the chart
plt.title('Distribution of Median House Value')
plt.xlabel('Median House Value')
plt.ylabel('Count')

# Display the genereated chart
plt.show()

```

```python
# Set the figure size
plt.figure(figsize=(10, 6))

#Create a scatter plot using 'median_income' for the X-axis and 'median_house_value' for the Y-axis
plt.scatter(df['median_income'], df['median_house_value'], alpha=0.3)

# Set the title and axis labels
plt.title('House Value vs. Income')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')

# Add grid lines for better readability
plt.grid(True)

# Display the plot
plt.show()
```
